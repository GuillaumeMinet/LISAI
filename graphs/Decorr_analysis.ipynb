```python
import csv
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path
```

```python
path = r"\\storage3.ad.scilifelab.se\testalab\Guillaume\01_Projects\DL_monalisa\_paper\Denoising_charac\Model_comp\DecorrelationAnalysis.csv"
```

```python
res = []
with open(path, 'r') as file:
    reader = csv.reader(file, delimiter=',')  # Use '\t' as the delimiter for tab-separated ys
    next(reader)  # Skip the first row (titles)
    for row in reader:
            res.append(float(row[2]))
print(res)
```

```python
pxsize = 30 # in nm

n_imgs = 3
labels = ["GT","CARE","HDNunsup","HDNsup","UNet-RCAN","SN2N","N2V"]

res_ = np.array(res)
res_ = np.reshape(res,(n_imgs,len(labels)),order='F')
print(res_)
```

```python
avg_res = np.mean(res_,axis=0)
print(avg_res)
std_res = np.std(res_,axis=0)
print(std_res)
```
