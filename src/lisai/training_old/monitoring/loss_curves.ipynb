```python
import os
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

# Parent directory containing the subfolders
parent_folder = r"E:\dl_monalisa\Models\Vim_fixed_mltplSNR_30nm\Upsampling_final_refinement"

# List all subfolders in the parent directory
subfolders = [f.path for f in os.scandir(parent_folder) if f.is_dir()]

# Initialize figure for loss curve
plt.figure(figsize=(12,6))

# Loop through each subfolder and read the loss.txt file
for folder in subfolders:
    loss_file = Path(folder) / "loss.txt"
    
    # Check if the file exists in the subfolder
    if loss_file.exists():
        # Load the data from the loss.txt file
        data = np.loadtxt(loss_file, skiprows=1)
        
        # Get a short name for the folder (last part of the path)
        folder_name = os.path.basename(folder)
        if len(folder_name) > 20:  # Customize this length threshold as needed
            folder_name = folder_name[:30] + "..."  # Truncate the folder name for brevity
        
        # Plot the training loss (assuming the structure is the same as the original code)
        plt.plot(data[:, 0], data[:, 1], label=f"{folder_name}")
# Finalize the training loss plot with an external legend
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')  # Place the legend outside the plot
plt.title("Training Loss Curves")
plt.tight_layout()  # Adjust layout to avoid clipping
plt.show()
```
